```python
# Create database named 'proj_db'
db = client['proj_db']


# Removes all existing collections
# db.collection.remove()
```

```python

# Create collection called 'data_collection'

collection_proj = db['data_collection']
```

```python
# Read json file into mongoDB
with open('new_json_file') as f:
    file_data = json.load(f)
    collection_proj.insert_one(file_data) 
    
client.close()
```
